```python
randomkey="LXNHXLLTXMVWPM-UHFFFAOYSA-N"
randomrow=0
def searchbykey(df,key):
    first_row = df.loc[key]
    selected_columns = first_row[first_row != 0.0]
    filtered_df = df[selected_columns.index]
    print(filtered_df.loc[[key]])

def searchbyrow(df,row):
    first_row = df.iloc[row]
    selected_columns = first_row[first_row != 0.0]
    filtered_df = df[selected_columns.index]
    print(filtered_df.iloc[[row]])
```

```python
searchbyrow(dfo,randomrow)
searchbyrow(dfn,randomrow)
```

```python
searchbykey(dfo,randomkey)
searchbykey(dfn,randomkey)
```
