```python
spectrums=[]
from matchms.importing import load_from_msp
spectrums = list(load_from_msp('../data/processed/MS2MassBank/MassBank_NIST_Feb20_metadata_toxfilter_peak_combined.msp'))
```

```python
import numpy as np
peak_dict = {}
for s in spectrums:
    inchikey = s.get('inchikey')
    peaks = np.vstack((s.peaks.mz, s.peaks.intensities)).T
    peak_dict[inchikey] = peaks
```

```python
peak_dict['CBGUOGMQLZIXBE-XGQKBEPLSA-N']
```

```python
inchikeys_MB=[]
for s in spectrums:
    inchikeys_MB.append(s.get('inchikey'))
#inchikeys_MB=set(inchikeys_MB)
inchikeys_MB=list(dict.fromkeys(inchikeys_MB))
len(inchikeys_MB)
```

```python
# NB! Here the index was obtained from set(inchikeys_MB). This has been changed to list(dict()), as the previous function changes the rows order (same content) of the intensity matrix (below).
# 2023_05_07_intensity_matrix_endpoints.csv was first used.
```

```python
# Creating an intensity matrix

import pandas as pd
# Create the table of bins
bin_edges = np.arange(10, 1000, 0.1)
bin_centers = (bin_edges[1:] + bin_edges[:-1]) / 2
n_bins = len(bin_centers)

# Initialize the intensity matrix
intensity_matrix = np.zeros((len(inchikeys_MB), n_bins))

# Assign the intensities to the corresponding bins
for i, inchi in enumerate(inchikeys_MB):
    peaks = peak_dict[inchi]
    for j in range(len(peaks)):
        mz = peaks[j][0]
        intensity = peaks[j][1]
        bin_idx = int((mz - 10) / 0.1)
        if bin_idx >= 0 and bin_idx < n_bins:
            intensity_matrix[i, bin_idx] = max(intensity_matrix[i, bin_idx], intensity)

# Create a pandas dataframe from the intensity matrix
df_peaks = pd.DataFrame(intensity_matrix, index=inchikeys_MB, columns=bin_centers)
# Remove bins without intensities
zero_cols = df_peaks.columns[(df_peaks == 0).all()]
df_peaks = df_peaks.drop(columns=zero_cols)
```

```python
df_endpoints = pd.read_csv('../data/processed/tox21/tox.csv', index_col='inchikey')
df = df_peaks.join(df_endpoints, on=df_peaks.index)
df.to_csv('../data/processed/MS2_tox/combined_spectra_vector_endpoints.csv')
```
